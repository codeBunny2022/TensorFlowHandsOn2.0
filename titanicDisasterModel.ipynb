```python
CATEGORICAL_COLUMNS=['Sex','Embarked']
NUMERICAL_COLUMNS=['Pclass','Age','SibSP','Parch','Fare']
```

```python
feature_columns=[]
for feature_name in CATEGORICAL_COLUMNS:
  vocabulary=dftrain[feature_name].unique() #gets list of all unique values
  feature_columns.append(tf.feature_column.categorical_column_with_vocabulary_list(feature_name,vocabulary))

for feature_name in NUMERICAL_COLUMNS:
  feature_columns.append(tf.feature_column.numeric_column(feature_name,dtype=tf.float32))

print(feature_columns)
```

```python
def make_input_fn(data_df,label_df,num_epoches=1,shuffle=False,batch_size=32):
  def input_function():
    ds=tf.data.Dataset.from_tensor_slices((dict(data_df),label_df))
    if shuffle:
      ds=ds.shuffle(1000)
    ds=ds.batch(batch_size).repeat(num_epoches)
    return ds
  return input_function

train_input_fn=make_input_fn(dftrain,y_train)
eva_input_fn=make_input_fn(dfeval,y_eval,num_epoches=1,shuffle=False)
```
